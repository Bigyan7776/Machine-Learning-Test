# (Question Number 1)

The total number of multiplications is k*m*n as the resulted matrix after multiplication is (n,m) and each element of the resulted matix is obtained by multiplying a row of the first matrix with a column of the second matrix componentwise.

And the total number of additions is (k-1)*m*n as the resulted matrix after multiplication is (n,m) and each element of the resulted matrix is obtained by adding the k number of products ,i.e., k-1 additions for each element of the resulted matrix.

For example: If A=[[1,2],[3,4]] and B=[[5,6],[7,8]],then the first element of AB is obtained by multiplying first row of A and the first column of B which has two multiplications and one addition. And since the size of AB is 2x2, there are in total 2*2*2 multiplications and 1*2*2 additions.

```python
# (Question Number 2)

# Import necessary libraries
import numpy as np
import time

# Define two matrices as lists of lists
matrix_a = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

matrix_b = [
    [9, 8, 7],
    [6, 5, 4],
    [3, 2, 1]
]

# Function to multiply matrices using lists
def multiply_matrices_lists(mat_a, mat_b):
    result = [[0 for _ in range(len(mat_b[0]))] for _ in range(len(mat_a))]

    for i in range(len(mat_a)):
        for j in range(len(mat_b[0])):
            for k in range(len(mat_b)):
                result[i][j] += mat_a[i][k] * mat_b[k][j]

    return result

# Function to multiply matrices using NumPy
def multiply_matrices_numpy(mat_a, mat_b):
    return np.dot(mat_a, mat_b)

# Timing the list multiplication
start_time_lists = time.time()
result_lists = multiply_matrices_lists(matrix_a, matrix_b)
end_time_lists = time.time()

# Timing the NumPy multiplication
start_time_numpy = time.time()
result_numpy = multiply_matrices_numpy(matrix_a, matrix_b)
end_time_numpy = time.time()

# Calculate time differences
time_taken_lists = end_time_lists - start_time_lists
time_taken_numpy = end_time_numpy - start_time_numpy

# Compare the results
if np.array_equal(result_lists, result_numpy):
    print("Results are the same.")
else:
    print("Results differ.")

# Compare the timings
if time_taken_lists < time_taken_numpy:
    print("Multiplication using lists is faster.")
    reason = "NumPy introduces some overhead for small matrices, and its optimized operations become more noticeable with larger matrices."
else:
    print("Multiplication using NumPy is faster.")
    reason = "NumPy is highly optimized for matrix operations and performs better, especially with larger matrices."

# Provide the timing details and reason
print(f"Time taken using lists: {time_taken_lists:.6f} seconds")
print(f"Time taken using NumPy: {time_taken_numpy:.6f} seconds")
print(f"Reason for the speed difference: {reason}")


        
```

# (Question Number 4)

The gradient of the function (x^2)y+(y^3)sin(x) with repect to x and y is a vector (2xy+(y^3)Cosx,x^2+3(y^2)Sin(x)).

```python
import jax
import jax.numpy as np

# Define the function
def f(x, y):
    return x**2 * y + y**3 * np.sin(x)

# Define the analytical gradients
def analytical_gradients(x, y):
    df_dx = 2 * x * y - y**3 * np.cos(x)
    df_dy = x**2 + 3 * y**2 * np.sin(x)
    return df_dx, df_dy

# Use JAX to calculate gradients
grad_f = jax.grad(f)

# Random values for x and y
random_x = np.random.rand()
random_y = np.random.rand()

# Evaluate gradients
analytical_grads = analytical_gradients(random_x, random_y)
jax_grads = grad_f(random_x, random_y)

# Compare the results
print("Analytical Gradients:", analytical_grads)
print("JAX Gradients:", jax_grads)
```

```python
from sympy import symbols, diff, sin

# Define symbols
x, y = symbols('x y')

# Define the function
f = x**2 * y + y**3 * sin(x)

# Calculate the partial derivatives
df_dx = diff(f, x)
df_dy = diff(f, y)

# Print the analytical gradients
print("Analytical Gradients:")
print("df/dx =", df_dx)
print("df/dy =", df_dy)
```

```python
# (Question Number 9)

import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0.5,100.5,0.5)
y1=x
y2=x**2
y3=x**3/100
y4=np.sin(x)
y5=(np.sin(x))/x
y6=np.log(x)
y7=np.exp(x)
# Create subplots
fig, axs = plt.subplots(1, 7, figsize=(30, 12))

# Plot the data on each subplot
axs[0].plot(x, y1, label='y=x')
axs[0].set_title('Graph 1')

axs[1].plot(x, y2, label='y=x^2')
axs[1].set_title('Graph 2')

axs[2].plot(x, y3, label='y=x^3/100')
axs[2].set_title('Graph 3')

axs[3].plot(x, y4, label='y=sin(x)')
axs[3].set_title('Graph 4')

axs[4].plot(x, y5, label='y=sin(x)/x')
axs[4].set_title('Graph 5')

axs[5].plot(x, y6, label='y=log(x)')
axs[5].set_title('Graph 6')

axs[6].plot(x, y7, label='y=e^x')
axs[6].set_title('Graph 7')


# Add common labels and title
for ax in axs:
    ax.set_xlabel('x-axis')
    ax.set_ylabel('y-axis')
    ax.legend()

plt.suptitle('Multiple Graphs Side by Side')

# Display the plots
plt.show()
```
